```python
#Frontend Script
from tkinter import *


#These function will call the view function of the backend.py file accept the result and store result in the list widget 
#of frontend.py
#These functions are important as some of them send parameters and accept result hence we can't achieve this 
#functionality by calling backend function through button

#This function will execute whenever an event occurs to which it is binded
def get_selected_row(event):
    global selected_tuple
    index = list1.curselection()[0] #Because we need to catch the first element of the tuple.
    selected_tuple = list1.get(index) #Retrieve the values of selected list record.

    e1.delete(0,END) #Clear existing data
    e1.insert(END,selected_tuple[1]) #Adding values of the selected tuple in the field(Entry tag)
    e2.delete(0,END)
    e2.insert(END,selected_tuple[2])
    e3.delete(0,END)
    e3.insert(END,selected_tuple[3])
    e4.delete(0,END)
    e4.insert(END,selected_tuple[4])


def view_command():
    list1.delete(0,END) #Clearing the List Widget when its executed again.
    for row in view():
        list1.insert(END,row) #END - New rows will be added at the end of the list box. 
        
def search_command():
    list1.delete(0,END)
    title = title_text.get()
    author = author_text.get() 
    year = year_text.get()
    isbn = isbn_text.get()
    
    for row in search(title,author,year,isbn):
        list1.insert(END,row)
        
def delete_command():
    delete(selected_tuple[0]) #To find the index of the current list highlighted row.
    view_command()
        
        
def insert_command():
    title = title_text.get()
    author = author_text.get() 
    year = year_text.get()
    isbn = isbn_text.get()
    insert(title,author,year,isbn)
    view_command()
    
def update_command():
    update(selected_tuple[0],title_text.get(),author_text.get() ,year_text.get(),isbn_text.get())
    view_command()
    
def clear():
    e1.delete(0,END)
    e2.delete(0,END)
    e3.delete(0,END)
    e4.delete(0,END)
    


window = Tk()
window.wm_title("BookStore")

#Static Labels
l1= Label(window,text="Title")
l1.grid(row=0,column=0)

l2 = Label(window,text="Author")
l2.grid(row=0,column=2)

l3=Label(window,text="Year")
l3.grid(row=1,column=0)

l4 = Label(window,text="ISBN")
l4.grid(row=1,column=2)

#Entries(Input Box)

title_text = StringVar()
e1 = Entry(window,textvariable=title_text)
e1.grid(row=0,column=1)

author_text = StringVar()
e2 = Entry(window,textvariable=author_text)
e2.grid(row=0,column=3)

year_text = StringVar()
e3 = Entry(window,textvariable=year_text)
e3.grid(row=1,column=1)

isbn_text = StringVar()
e4 = Entry(window,textvariable=isbn_text)
e4.grid(row=1,column=3)


#Listbox

list1 = Listbox(window,height=8,width=45)
list1.grid(row=2,column=0,rowspan=8,columnspan=2)

#Scrollbar

sb1 = Scrollbar(window)
sb1.grid(row=2,column=2,rowspan=8)

#Linking ListBox to scrollbar

list1.configure(yscrollcommand = sb1.set)
sb1.configure(command=list1.yview)

#Trigger Management for an listbox event
list1.bind('<<ListboxSelect>>',get_selected_row) #Binding an event('Clicking on a record of a list box to a particular finction'),

#Adding Buttons

b1 = Button(window,text="View All",width=15,command=view_command)
b1.grid(row=2,column=3)

b2 = Button(window,text="Search Entry",width=15,command=search_command)
b2.grid(row=3,column=3)

b3 = Button(window,text="Add Entry",width=15,command=insert_command)
b3.grid(row=4,column=3)

b4 = Button(window,text="Update Entry",width=15,command=update_command)
b4.grid(row=5,column=3)

b5 = Button(window,text="Delete Entry",width=15,command=delete_command)
b5.grid(row=6,column=3)


b6 = Button(window,text="Close ",width=15,command=window.destroy)
b6.grid(row=7,column=3)

b7 = Button(window,text="Clear",width=15,command=clear)
b7.grid(row=8,column=1)



window.mainloop()

connect()
```

```python
#Backend script

import sqlite3

def connect():
    con=sqlite3.connect("library.db")
    cur=con.cursor()
    if(cur.execute("CREATE TABLE IF NOT EXISTS book (id INTEGER PRIMARY KEY,title TEXT,author TEXT,year INTEGER, isbn INTEGER)")):
        print('Table Created')
    con.commit()
    con.close()
    
def insert(title,author,year,isbn):
    con=sqlite3.connect("library.db")
    cur=con.cursor()
    cur.execute("Insert into book VALUES(NULL,?,?,?,?)",(title,author,year,isbn))
    con.commit()
    con.close()

def view():
    con=sqlite3.connect("library.db")
    cur=con.cursor()
    cur.execute("Select * from book")
    rows = cur.fetchall()
    con.commit()
    return(rows) 
    con.close()
    
def search(title="",author="",year="",isbn=""):
    con=sqlite3.connect("library.db")
    cur=con.cursor()
    cur.execute("Select * from book where title = ? or author = ? or year = ? or isbn = ?",(title,author,year,isbn))
    rows = cur.fetchall()
    con.commit()
    con.close()
    return rows
    
def update(id,title,author,year,isbn):
    con=sqlite3.connect("library.db")
    cur=con.cursor()
    cur.execute("Update book set title=?,author=?,year=?,isbn=? where id=?",(title,author,year,isbn,id))
    con.commit()
    con.close()

def delete(id):
    con=sqlite3.connect("library.db")
    cur=con.cursor()
    cur.execute("Delete from book where id=?",(id,))
    con.commit()
    con.close()
    
    
```
